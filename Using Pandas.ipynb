### Example of geoaddressing with Geopy and Pandas

```python
from geopy.geocoders import ArcGIS
```

```python
nom = ArcGIS()
```

```python
type(nom)
```

```python
df1
```

```python
df1["Address"] = df1["Address"]+", "+df1["City"]+", "+df1["State"]+", "+df1["Country"]
```

```python
df1
```

```python
df1["Address"]
```

```python
df1["Coordinates"] = df1["Address"].apply(nom.geocode)
```

**Note:** *apply() is a special pandas function which calls a function on all the objects of the series datatype Address , i.e. saving us from having to run a loop*

```python
df1
```

**If you prefer to do it without using the** apply() **method:** 

```python
lis = []
for add in df1["Address"]:
    lis.append(nom.geocode(add))
df1["Coordinates"] = lis
```

```python
df1
```

###### See how the latitude and longitude can now be accessed:

```python
print(df1["Coordinates"][0].latitude)
print(df1["Coordinates"][0].longitude)
```

```python
df1["Coordinates"].apply(lambda x:x.latitude)
```

*Due to longer string length the coordinates are displayed using ... at the end in the DataFrame*

**Lets make seperate columns for the latitude and the longitude**

```python
df1["latitude"] = df1["Coordinates"].apply(lambda x:x.latitude)
df1["longitude"] = df1["Coordinates"].apply(lambda x:x.longitude)
```
